```python

# Group by 'District' and calculate max, min, and mean for each column
district_humidity = df.groupby('District')['Humidity_2m'].max().reset_index()
district_temp_max = df.groupby('District')['MaxTemp_2m'].max().reset_index()
district_temp_min = df.groupby('District')['MinTemp_2m'].min().reset_index()
district_temp_mean = df.groupby('District')['Temp_2m'].mean().reset_index()
district_precip = df.groupby('District')['Precip'].sum().reset_index()  # Total precipitation
district_pressure = df.groupby('District')['Pressure'].mean().reset_index()
district_wind_10m = df.groupby('District')['WindSpeed_10m'].mean().reset_index()
district_wind_50m = df.groupby('District')['WindSpeed_50m'].mean().reset_index()
```

```python
district_map = district_temp_max.copy()  # Start with one DataFrame

# Merge all data based on 'District'
district_map = district_map.merge(district_temp_min, on="District")
district_map = district_map.merge(district_temp_mean, on="District")
district_map = district_map.merge(district_humidity, on="District")
district_map = district_map.merge(district_precip, on="District")
district_map = district_map.merge(district_pressure, on="District")
district_map = district_map.merge(district_wind_10m, on="District")
district_map = district_map.merge(district_wind_50m, on="District")

# Display the final dataset
district_map
```

```python
district_location = df[['District', 'Latitude', 'Longitude']].drop_duplicates()
district_map = pd.merge(district_map, district_location, on="District")
```

```python
df1 = df.copy()
df1['Date'] = pd.to_datetime(df1['Date'])
df1.head()
```

```python
# Create new features like day of the year, month, or hour if necessary
df1['DayOfYear'] = df1['Date'].dt.dayofyear
df1['Month'] = df1['Date'].dt.month
df1['Year'] = df1['Date'].dt.year
df1.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2 = df1.copy()
df2['District'] = le.fit_transform(df['District'])
df2.head()
```

```python
features = ['Precip', 'Pressure', 'Humidity_2m', 'WindSpeed_10m', 'MaxWindSpeed_10m', 'MinWindSpeed_10m', 'Temp_2m', 'DayOfYear', 'Month', 'Year']
X = df2[features]  # Features
y = df2['MaxTemp_2m']  # Target variable
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
X_train.shape,X_test.shape
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error,mean_absolute_error, r2_score
import xgboost as xgb
import lightgbm as lgb
from catboost import CatBoostRegressor
```

```python
models = {
    'LinearRegression': LinearRegression(),
    'Ridge': Ridge(),
    'Lasso': Lasso(),
    'ElasticNet': ElasticNet(),
    'DecisionTree': DecisionTreeRegressor(),
    'RandomForest': RandomForestRegressor(),
    'GradientBoosting': GradientBoostingRegressor(),
    'XGB': xgb.XGBRegressor(),
    'LGBM': lgb.LGBMRegressor(),
    'CatBoost': CatBoostRegressor(silent=True)  # Disable output for CatBoost
}
```

```python
param_grid = {
    'LinearRegression': {},
    'Ridge': {
        'alpha': [0.1, 1, 10, 100]
    },
    'Lasso': {
        'alpha': [0.1, 1, 10, 100]
    },
    'ElasticNet': {
        'alpha': [0.1, 1, 10],
        'l1_ratio': [0.1, 0.5, 0.9]
    },
    'DecisionTree': {
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4]
    },
    'RandomForest': {
        'n_estimators': [50, 75, 100],
        'max_depth': [None, 10, 20],
        'min_samples_split': [2, 5, 10]
    },
    'GradientBoosting': {
        'n_estimators': [50, 75, 100],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 10]
    },
    'XGB': {
        'n_estimators': [50, 100, 75],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 10]
    },
    'LGBM': {
        'n_estimators': [50, 100, 75],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 5, 10]
    },
    'CatBoost': {
        'iterations': [50, 100, 75],
        'learning_rate': [0.01, 0.1, 0.2],
        'depth': [3, 5, 10]
    }
}
```

```python
# Perform GridSearchCV for each model
best_models = {}
for model_name, model in models.items():
    print(f"Training and tuning {model_name}...")
    
    # Perform GridSearchCV
    grid_search = GridSearchCV(estimator=model, param_grid=param_grid[model_name], cv=3, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
    grid_search.fit(X_train, y_train)
    
    best_models[model_name] = grid_search.best_estimator_
    print(f"                                   Best parameters for {model_name}: {grid_search.best_params_}")
    
    # Predict with the best model
    y_pred = grid_search.best_estimator_.predict(X_test)
    
    # Evaluate the model using multiple metrics
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)  # RMSE is the square root of MSE
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print("-"*130)
    print(f"Evaluation metrics for {model_name}:")
    print(f"  MSE: {mse}")
    print(f"  RMSE: {rmse}")
    print(f"  MAE: {mae}")
    print(f"  R²: {r2}")
    print("-"*130)
    
    
    # Plotting only the first 100 points to avoid clutter
    plt.figure(figsize=(12, 6))
    plt.plot(y_test.values[:100], label='True Temperature', color='blue', linewidth=2)
    plt.plot(y_pred[:100], label='Predicted Temperature', linestyle='--', color='orange', linewidth=2)
    plt.xlabel('Sample Index')
    plt.ylabel('Max Temperature (°C)')
    plt.title(f'{model_name} - Predicted vs Actual Max Temperature')
    plt.legend()
    plt.show()
```
